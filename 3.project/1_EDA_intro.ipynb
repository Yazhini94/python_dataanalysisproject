```python
import ast
# Importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading dataset
datasets = load_dataset('lukebarousse/data_jobs')   # Correct source name and variable name
df = datasets['train'].to_pandas()                  # Corrected: 'datasets' instead of 'dataset'

# Data Cleanup
#  # Optional: Check actual column names
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) 
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x) # Use correct column name
```

#### Filter for US Data Analyst

```python
df_DA_US=df[(df['job_country']== 'United States') & (df['job_title_short']=='Data Analyst')]
```

```python
import seaborn as sns
print(sns.__version__)
```

```python
df_plot=df_DA_US['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data= df_plot,x='count',y='job_location',hue='count',palette='dark:b_r')
sns.despine()
plt.title('Number of jobs per job Title')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```

```python
dict_column={
    'job_work_from_home':'work from Home offered',
    'job_no_degree_mention':'Degree Requirement',
    'job_health_insurance':'Health Insurance Offered'
}

fig,ax=plt.subplots(1,3)
fig.set_size_inches((12,5))

for i,(column,title)in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(),labels=['False','True'],autopct='%1.1f%%',startangle=90)
    ax[i].set_title(title)

plt.show()
```

```python
df_plot=df_DA_US['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data= df_plot,x='count',y='company_name',hue='count',palette='dark:b_r')
sns.despine()
plt.title('Number of companies Data Analyst in the US')
plt.xlabel('Number of companies')
plt.ylabel('')
plt.show()
```
